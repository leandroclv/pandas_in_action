```python
employees = pd.read_csv('employees.csv', parse_dates=['Start Date'])
employees
```

```python
employees.dropna() #o método dropna() deletou todas as linhas que possuiam valores NaNs do dataframe
```

```python
employees.dropna(how='all').tail() #o parâmetro how recebe o argumento 'all' para remover todas as linhas que possuem todos os valores NaN. Por padrão o parâmetro how possue o argumento 'any' que remove as linhas que possue pelo menos um valor NaN
```

```python
employees.dropna(how='any').tail() #Note que na célula passada a linha de índice 995 possui um valor NaN na coluna Gender e irá sumir com o parâmetro how = 'any'
```

```python
employees.dropna(subset=['Gender']).tail() #Podemos usar o parâmetro 'subset' para linhas com valores NaN em colunas específicas
```

```python
employees.dropna(subset=['Start Date', 'Salary']).head() #Podemos passar para o parâmetro subset uma lista com colunas para restirar as linhas que possuem valores NaN 
```

```python
employees['Team'].head()
```

```python
employees['Team'].duplicated().head() #mesma coisa employees['Team].duplicate(keep = 'first').head()

#O método duplicated retorna uma serie booleana que identifica duplicatas nas colunas. O método marca a primeira ocorrência como não duplicado (False), todos os outros valores duplicados depois serão marcados como True
```

```python
employees['Team'].duplicated(keep='last') #Marcar as últimas ocorrências de valores duplicadas
```

```python
(~employees['Team'].duplicated()).head() #inverte 
```

```python
first_one_in_team = ~employees['Team'].duplicated()
employees[first_one_in_team]
```

```python
employees.drop_duplicates()
```

```python
employees.drop_duplicates(subset=['Team']) #Retorna quais índices apareceu pela primeira vez os valores únicos da coluna Team
```

```python
employees.drop_duplicates(subset=['Team'], keep='last') #Retorna quais índices apareceu pela última vez os valores únicos da coluna Team
```

```python
employees.drop_duplicates(subset=['First Name'], keep=False) #Retorna quais nomes apareceram pela primeira vez os valores únicos da coluna First Name. O argumento False irá excluir todas as linhas com valores duplicados
```

```python
name_is_doulgas = employees['First Name'] == 'Douglas' #usando a serie bolleana para armazenar na variável as linhas que possuem a palavra Doulgas na coluna First Name
is_male = employees['Gender'] == 'Male'  #usando a serie bolleana para armazenar na variável as linhas que possuem a palavra Male na coluna Gender
employees[name_is_doulgas & is_male] #Retorna as linhas que possuem Doulgas na coluna First Name e Male na coluna Gender
```

```python
employees.drop_duplicates(subset=['Gender', 'Team']).head() #Retorna todos os valores que possuem duplicação na coluna Gender e First Name. Caso uma das duplicadas seja falso, retornará, por exemplo: First Name Douglas e Gender Male, irá aparecer se o Fisrt Name Douglas e Gender NaN (que é diferente de Male)
```
