```python
import pandas as pd
import numpy as np
```

```python
df_employees = pd.read_csv('./employees.csv', parse_dates=['Start Date'])
df_employees
```

```python
df_employees.info()
```

```python
df_employees['Mgmt'].astype(bool) #Transformou a coluna Mgmt de tipo object para bool
```

```python
df_employees['Mgmt'] = df_employees['Mgmt'].astype(bool) #Transformando o tipo de dado da coluna Mgmt para booleano
```

```python
df_employees.info()
```

```python
df_employees['Salary'].fillna(0).tail() #Transformou os falores NaN da coluna Salary para 0
```

```python
df_employees['Salary'].fillna(0).astype(int).tail()
```

```python
df_employees['Salary'] = df_employees['Salary'].fillna(0).astype(int)
```

```python
df_employees.info()
```

```python
df_employees.nunique() #Mostr quantos valores únicos o DataFrame possui
```

```python
df_employees['Gender'].astype('category') #Transforma o tipo de dado desta coluna para category
```

```python
df_employees['Gender'] = df_employees['Gender'].astype('category') #salva a transformação
```

```python
df_employees.info()
```

```python
df_employees['Team'].astype('category') #Transforma o tipo de dado desta coluna para category
```

```python
df_employees['Team'] = df_employees['Team'].astype('category') #salva a transformação
```

```python
df_employees.info()
```

```python
df_employees['First Name'] == 'Maria' #Verifica em toda a serie se existe a palavra Maria, caso exista retorna True se não False. Chamado serie booleana percorrendo toda a coluna e classificando True ou False para a operação que foi pedida (== 'Maria')
```

```python
df_employees[df_employees['First Name'] == 'Maria'] #utilizando a serie booleana retorna as linhas que deram True
```

```python
marias = df_employees['First Name'] == 'Maria' #Este código percorre a serie e os dados que possuem True para a comparação dataframe['coluna] == 'condição' e a linha é adiconada a variável marias
df_employees[marias]
```

```python
df_employees['Team'] != 'Finance'
```

```python
df_employees[df_employees['Team'] != 'Finance'] #utilizando a serie booleana retorna as linhas que deram False
```

```python
high_earners = df_employees['Salary'] > 100000
high_earners.head()
```

```python
df_employees[df_employees['Salary'] > 100000].head()
```

```python
is_female = df_employees['Gender'] == 'Female'
is_female
```

```python
in_biz_dev = df_employees['Team'] == 'Business Dev'
in_biz_dev
```

```python
df_employees[is_female & in_biz_dev].head() #Interceção das duas series utlizando o operador lógico and (&)
```

```python
is_manager = df_employees['Mgmt']
df_employees[is_female & in_biz_dev & is_manager].head() #Podemos utilizar duas ou mais series utilizando o and
```

```python
earning_below_40k = df_employees['Salary'] < 40000
earning_below_40k
```

```python
started_after_2015 = df_employees['Start Date'] > '2015-01-01'
started_after_2015
```

```python
df_employees[earning_below_40k|started_after_2015].head() #Interceção das duas series utlizando o operador lógico or (|), o pandas coloca também as linhas que possuem True e True também
```

```python
df_employees[~ (df_employees['Salary'] < 40000)].head() #Interceção da serie utlizando o operador lógico not (~), aqui vou invertido para as linhas que recebem mais de 40000
```

```python
sales = df_employees['Team'] == 'Sales'
legal = df_employees['Team'] == 'Legal'
mktg = df_employees['Team'] == 'Marketing'
df_employees[sales | legal | mktg].head()
```

```python
all_star_teams = ['Sales','Legal','Marketing']  #Utilizado quando se deseja fazer muitas comparações com o or
on_all_star_teams = df_employees['Team'].isin(all_star_teams)
df_employees[on_all_star_teams].head()
```

```python
between_80K_and_90K = df_employees['Salary'].between(80000, 90000) #Foi usado o método 'between' na coluna Salary para mostras as pessoas que ganham de 80.000 até 90.000. Este resultado é (includente:excludente)
df_employees[between_80K_and_90K].head()
```

```python
eighties_folk = df_employees['Start Date'].between(left='1980-01-01', right='1990-01-01') #between pode ser usado para datas também
df_employees[eighties_folk].head()
```

```python
name_starts_with_r = df_employees['First Name'].between('R', 'S') #between pode ser usado para strings também
df_employees[name_starts_with_r].head()
```
