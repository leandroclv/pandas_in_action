```python
import pandas as pd
import numpy as np
```

```python
df_inspections = pd.read_csv('chicago_food_inspections.csv')
df_inspections.head()
```

```python
df_inspections['Name'].head()
```

```python
df_inspections['Name'].head().values
```

```python
df_inspections['Name'].str.strip().head() #O método strip() retira os espaços antes e depois do texto, como no Python
```

```python
df_inspections['Name'] = df_inspections['Name'].str.strip()
```

```python
df_inspections
```

```python
df_inspections['Name'].str.lower().head() #O método lower() transforma todo o texto em minúsculo, como no Python
```

```python
df_inspections['Name'].str.upper().head() #O método upper() transforma todo o texto em maiúscula, como no Python
```

```python
df_inspections['Name'].str.capitalize().head() #O método capitalize() transforma toda 1° letra do texto em maiúscula, como no Python
```

```python
df_inspections['Name'].str.title().head() #O método title() transforma todo o começo do texto em maiúscula, como no Python
```

```python
df_inspections['Name'] = df_inspections['Name'].str.title()
df_inspections.head()
```

```python
df_inspections['Risk'].head()
```

```python
df_inspections['Risk'].unique() #Mostra quais os valores únicos do DataFrame, note que existem valores NaN e all
```

```python
df_inspections = df_inspections.dropna(subset=['Risk']) #Foram retirados os valores NaN da coluna Risk
```

```python
df_inspections['Risk'].unique()
```

```python
df_inspections = df_inspections.replace(to_replace='All', value='Risk 4 (Extreme)') #Transformou o valor 'All' em 'Risk 4 (Extreme)'
```

```python
df_inspections['Risk'].unique()
```

```python
df_inspections['Risk'].str.slice(5,6).head() #Mostra do índíce 5 no texto até o 6 [5:6] -> [inclusivo:exclusivo]
```

```python
df_inspections['Risk'].str[8:-1].head() #outra forma de usar o slice mostrando do índice 8 no texto até o último índice
```
