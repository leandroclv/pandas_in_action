```python
buckets = [0, 200, 400, 600, 800, 1000, 1200, 1400]
df_google_stocks.value_counts(bins=buckets).sort_index() 
#Foi feito uma lista de 0 a 1400 com instervalos de 200 para classificar a diferença do menor valor para o maior dentro de cada intervalo. Entre os valores de 1200 e 1400 aparecem 93 vezes, por exemplo. A maioria dos valores estão entre 200 e 400 neste caso. O intervalo de captura de todos os valores maiores que, possui um '(' marcando um valor excludente do intervalo, como no caso de -0.001 e menores ou iguais, possui um ']' marcando um valor includente do intervalo, no caso do 200.
#O parâmetro bin retorna intervalos semiabertos, neste caso a lista com os valores estabelecidos na lista como guias
# Poderia escrever desse jeito também 'google.value_counts(bins = buckets, sort = False)'
```

```python
df_google_stocks.value_counts(bins=6, sort = False) #Dividiu em 6 grupos os valores e a ordem ficou do maior para o menor
```

```python
#Função criada para classificar Pokemons como multiplos ou únicos
def single_or_multi(pokemon_type):
    if '/' in pokemon_type:
        return 'multi'
    return 'Single'
```

```python
df_pokemon.apply(single_or_multi)
```

```python
df_pokemon.apply(single_or_multi).value_counts()
```

```python
days_of_wars = pd.read_csv(
    'revolutionary_war.csv',
    usecols=['Start Date'],
    parse_dates=['Start Date']
)
days_of_wars.squeeze()
```
