```python
import pandas as pd
import numpy as np
```

```python
pd.Series() #Criar uma série
```

```python
ice_cream_flavors = [
    'Chocolate',
    'Vanilla',
    'Strawberry',
    'Rum Raisin'
]
pd.Series(ice_cream_flavors) #Foi criado uma Série da variável ice_cream_flavors
```

```python
pd.Series(data=ice_cream_flavors) #O argumento data nesta função recebe a variável e os outros argumentos continuam no padrão
```

```python
days_of_week = ('Monday', 'Wednesday', 'Friday', 'Saturday')
```

```python
pd.Series(ice_cream_flavors, days_of_week) #Indicamos quem seriam os índices da Série
```

```python
pd.Series(data=ice_cream_flavors, index=days_of_week)
```

```python
days_of_week = ('Monday', 'Friday', 'Saturday', 'Monday')
```

```python
pd.Series(data=ice_cream_flavors, index=days_of_week) #Pandas permite duplicatas nos índices, contuto não é recomendado fazer para não ter problemas quando for preciso encontrar o valor deste índice com maior rapidez
```

```python
pd.Series(index=days_of_week, data=ice_cream_flavors) #Mesmo mudando a ordem dos parâmetros o pandas entende o que é index e data
```

```python
numbers = pd.Series([1,2,3,np.nan, 4, 5])
numbers
```

```python
numbers.count() #Conta a quantidade de linhas com valores
```

```python
numbers.sum() #Soma todos os valores da coluna
```

```python
#A maioria dos métodos matemáticos ignoram os dados perdidos (NaN) por padrão. Pode ser passado um argumento de False para skipna incluir esses valores nos métodos matemáticos.Retorna um nan, pois não pode ser adicionado a um valor desconhecido

numbers.sum(skipna=False)
```

```python
numbers.product() #Multiplica todo os valores da coluna
```

```python
numbers.cumsum() #Faz uma soma cumulativa para determinar qual valor contribui mais do que o total
```

```python
numbers.pct_change() #Retorna a porcentagem de diferença de um valor da série para outra
```

```python
numbers.pct_change(fill_method='bfill')
```

```python
numbers.mean() #Média
```

```python
numbers.median() #Mediana
```

```python
numbers.std() #Devio padrão (variação nos dados)
```

```python
numbers.min() #Valor mínino
```
