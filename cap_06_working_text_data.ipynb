```python
df_custumers = pd.read_csv('customers.csv')
df_custumers
```

```python
df_custumers['Name'].str.len().head() #O método str.len() retorna o número de caracteres dos valores da coluna Name
```

```python
df_custumers['Name'].str.split(' ').head() #O método str.split() irá separar o texto pelos espaços (' ') e colocará as palavras em uma lista
```

```python
df_custumers['Name'].str.split(' ').str.len().head() #O método str.split() irá separar o texto pelos espaços (' ') e colocará as palavras em uma lista e o método str.len() mostra o tamanho dos componentes da lista
```

```python
df_custumers['Name'].str.split(pat = ' ', n=1).head() #O método str.split() irá separar o texto pelos espaços (' ') e colocará as palavras em uma lista e o n=1 o número máximo de splits 
```

```python
df_custumers['Name'].str.split(pat=' ', n=1).str.get(0).head() #O método str.split() irá separar o texto pelos espaços (' ') e colocará as palavras em uma lista e o n=1 o número máximo de splits e o método str.get(0) seleciona o primeiro elemento (índice 0)
```

```python
df_custumers['Name'].str.split(pat=' ', n=1).str.get(1).head() #O método str.split() irá separar o texto pelos espaços (' ') e colocará as palavras em uma lista e o n=1 o número máximo de splits e o método str.get(1) seleciona o segundo elemento (índice 1). O método str.get() suporta números negativos (de trás para frente em relacão aos índices)
```

```python
df_custumers['Name'].str.split(pat=' ', n=1, expand= True).head() #O parâmetro expand = True cria um dataframe com duas colunas com os índices 0 e 1. Caso não coloque o parâmetro n=1 o Pandas vai expandir para todas as palavras e as que não tiverem ele coloca None no lugar
```

```python
df_custumers[['First Name', 'Last Name']] = df_custumers['Name'].str.split(pat=' ', n=1, expand=True) #Foram cradas duas colunas no DataFrame df_custumes e colocado os valores das colunas expendidas 
```

```python
df_custumers
```

```python
df_custumers = df_custumers.drop(labels='Name', axis='columns') #Foi excluído a coluna Name de DataFrame
df_custumers.head()
```
