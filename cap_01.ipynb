```python
df[df['Studio'] == 'Universal'] #Mostra todos os filmes que na coluna Studio possui o dado Universal
```

```python
released_by_universal = (df['Studio'] == 'Universal') #O resultado obtido na célula anterior foi adicionado a uma variável
df[released_by_universal].head()
```

```python
released_in_2015 = df['Year'] == 2015 #adicionou a uma variável os filmes com o dado 2015 na coluna Year
df[released_by_universal & released_in_2015] #Filtrou utilizando as variáveis os filmes que são da Universal e(&) do ano de 2015 
```

```python
df[released_by_universal | released_in_2015] #Filtrou utilizando as variáveis os filmes que são da Universal ou(|) do ano de 2015 
```

```python
before_1975 = df['Year'] < 1975 #Adicionou em uma variável os  filmes que possuem na coluna Year dados menores que 1975
df[before_1975]
```

```python
mid_80s = df['Year'].between(1983, 1986) #Adicionou em uma variável os filmes que possuem na coluna Year dados entre 1983 e 1986
df[mid_80s]
```

```python
has_dark_in_title = df.index.str.lower().str.contains('dark') #Adicionou em uma variável o filtro na coluna dos índices a palavra dark depois de passar todos os títulos para a forma minúscula 
df[has_dark_in_title]
```

```python
df['Gross'].str.replace('$', '', regex = False).str.replace(',', '', regex = False)
#Foi retirado da coluna Gross os símbolos $ e , (replace) para no próximo passo transformar os dados de object (string) para float64
```

```python
df['Gross'].str.replace('$', '', regex = False).str.replace(',', '', regex = False).astype(float)
#Foi transformado os dados da coluna Gross de object para float64 (astype(float))
```

```python
df['Gross'] = df['Gross'].str.replace('$', '', regex = False).str.replace(',', '', regex = False).astype(float)
#O pandas usa uma cópia para trabalhar no dados e não danificar o original. Neste passo foi modificado a coluna original Gross com os dados transformados
```

```python
df['Gross'].mean() #Como os dados da coluna Gross foram transformados em float dá para saber a média do valor total dos filmes
```

```python
studios = df.groupby('Studio') #Foi agrupado as linhas do DataFrame baseados nos valores na coluna Studio
```

```python
studios['Gross'].count().head() #Contar o número de filme por studio
```

```python
studios['Gross'].count().sort_values(ascending = False).head() #Filtrar do studio que possui a maior quantidade  de filmes para o menor na coluna Gross
```

```python
studios['Gross'].sum().head() #Somou a quantidade de valor da coluna Gross para cada studio
```

```python
studios['Gross'].sum().sort_values(ascending = False).head() #Filtrou do studio que possui o maior valor para o menor da coluna Gross
```
