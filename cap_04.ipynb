```python
df_nba.sample(3) #O método sample(n) coleta n números de linhas aleatórias do DataFrame 
```

```python
df_nba.nunique() #no DataFrame o método nunique conta o número de valores únicos presentes nem cada coluna, diferente da serie que este método retorna os valores únicos de toda a serie
```

```python
df_nba.max() #Diferente da serie que o método 'max()' retorna o valor máximo da serie, no DataFrame este método retorna valor máximo de cada coluna (em ordem de número, data e alfabética)
```

```python
df_nba.min() #Diferente da serie que o método 'min()' retorna o valor mínimo da serie, no DataFrame este método retorna valor mínimo de cada coluna (em ordem de número, data e alfabética)
```

```python
df_nba.nlargest(n=4, columns='Salary') #Este método 'nlargest()' encontra os maiores valores. Precisa como parâmetros a quantidades de valores desejados para encontrar e em que coluna procurar. Neste caso procuramos os 4 jogadores com os maiores salários
```

```python
df_nba.nsmallest(n=3, columns=['Birthday']) #Este método 'nsmallest()' encontra os menores valores. Precisa como parâmetros a quantidades de valores desejados para encontrar e em que coluna procurar. Neste caso procuramos os 3 jogadores mais velhos 
```

```python
df_nba.sum(numeric_only=True) #usando o método 'sum()' no DataFrame para descobir a soma de todos os salários, como visto na célula passada, retorna um erro. Por isso é utilizado o parâmetro numeri_only = True para ter a soma da coluna que possui somente números como tipo de dados
```

```python
df_nba.mean(numeric_only=True) #Média salarial de todos os jogadores 
```

```python
df_nba.median(numeric_only=True) #Mediana dos salários de todos os jogadores
```

```python
df_nba.mode(numeric_only=True) #Modo 
```

```python
df_nba.std(numeric_only=True) #Desvio padrão dos salários
```

```python
df_nba.sort_values(by = 'Name') #O método 'sort_values()' precisa de um parêmtro (nome da coluna) para pode ordenar
```

```python
df_nba.sort_values(by = 'Name', ascending=False) #Por padrão o parâmetro 'ascending' vem como True e a ordem é crescente, para inverter é só colocar False no ascending
```

```python
df_nba.sort_values(by='Birthday', ascending=False).head() #Outra maneira de encontrar os jogadores mais novos sem utilizar o método 'nsamllest()' 
```

```python
df_nba.sort_values(by=['Team', 'Name']) #O método 'sort_values()' pode utilizar mais de uma coluna como parâmetro para ordenar o DataFrame seguindo a ordem da lista de colunas. Por padrão o ascending vai estar em True (ordem crescente). Neste exemplo primeiro vai seguir a ordem alfabética da coluna Team e depois a ordem alfabética da coluna Name
```

```python
df_nba.sort_values(by = ['Team', 'Salary'],ascending=[True, False]) #O Team terá a ordem alfabética na ordem crescente e o Salary na ordem decrescente 
```

```python
df_nba = df_nba.sort_values(by = ['Team', 'Salary'],ascending=[True, False]) 
```

```python
df_nba.head() #Como foi salvo na célula passada a ordem dos dados o DataFrame ficou com os índices fora da ordem
```

```python
df_nba.sort_index().head() #Ordenou o DataFrame pelo índice. Por padrão o ascending está True e fica em ordem crescente
```

```python
df_nba = df_nba.sort_index()
df_nba.head()
```

```python
df_nba.sort_index(axis=1).head() #Ordenou os rótulos das colunas (axis=1) na ordem alfabética crescende. Por padrão o ascending vem True, caso queira colocar em ordem decrescente é só colocar o argumento ascending = False
```

```python
df_nba = pd.read_csv('nba.csv', parse_dates=['Birthday'], index_col='Name') #Quando formos criar um DataFrame podemos adicionar os parâmetros 'parse_dates' para mudar o tipo da coluna Birthday de string para datetime e o 'index_col' para colocar o índice como sendo a coluna Name 
```

```python
df_nba.head()
```

```python
df_nba['Salary'] #Seleciona a coluna Salary somente (pode ser também df_nba.Salary, mas este código não dá pra selecionar mais de uma coluna)
```

```python
df_nba[['Salary', 'Birthday']] #Lembrar que tem que ser uma lista para colocar como parâmetro, por isso [[]]
```

```python
df_nba.select_dtypes(include='object') #Seleciona as colunas que possuem o dtype object
```

```python
df_nba.select_dtypes(exclude=['object', 'int']) #Seleciona as colunas que não possuem o dtype object e int
```

```python
df_nba.loc['LeBron James'] #Seleciona a linha cujo o índice é o nome do jogador retornando uma serie dos dados da linha 
```

```python
df_nba.loc[['Kawhi Leonard', 'Paul George']] #No DataFrame o método 'loc' permite selecionar multiplas linhas e retorna um DataFrame organizado as linhas pela ordem da lista. o método 'loc' aceita um parâmetro podendo ser um dado ou uma lista de dados, por isso do [[]]
```

```python
df_nba.sort_index().loc['Otto Porter':'Patrick Beverley']#Utilizamos o 'sort_index' para organizar por índice (nomes dos jogadores) na ordem alfabética crescente e utilizamos o 'loc' para acessar os jogadores que estão entre o Otto Portter, menor valor, e o Patrick Beverley, maior valor -> ('Otto Porter':'Patrick Beverley'). Note que no slice do Pandas os dois valore são incluídos, diferente do Python que é [incluso:excluso]
```

```python
df_nba.sort_index().loc['Zach Collins':] #Faz um slice do Zach até o final 
```

```python
df_nba.sort_index().loc[:'Al Horford'] #Faz um slice do Al Horford até o começo
```
