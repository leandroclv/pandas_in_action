```python
addresses = [
("8809 Flair Square", "Toddside", "IL", "37206"),
("9901 Austin Street", "Toddside", "IL", "37206"),
("905 Hogan Quarter", "Franklin", "IL", "37206"),
]
```

```python
pd.MultiIndex.from_tuples(addresses) #Forma um MultiIndex onde o primeiro valor das tuplas é o endereço, o segundo valor é a cidade, o terceiro valor é o estado e o quarto valor é o código zip
```

```python
row_index = pd.MultiIndex.from_tuples(
    tuples = addresses,
    names = ['Street', 'City', 'State', 'Zip']
)
row_index
```
